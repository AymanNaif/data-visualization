```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Set seaborn

```python
sns.set()
```

## DataFrame

```python
df = sns.load_dataset('penguins')
df.head()
```

## Bar Plot for first 15 record

```python
sns.barplot(x='bill_length_mm', y='body_mass_g', data=df.head(15))
```

## Linear Regrusion with Plot

```python
sns.lmplot(x='bill_length_mm', y='body_mass_g', data=df)
```

## Relational Plot

```python
sns.relplot(x='bill_length_mm', y='body_mass_g', data=df)
```

## Summary

#### The more bill length , the greater body mass
