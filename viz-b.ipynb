```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
sns.boxplot(x='bill_length_mm', y='body_mass_g', data=df.head(50))
```

```python
sns.lmplot(x='bill_length_mm', y='body_mass_g', data=df)
```

```python
sns.relplot(x='bill_length_mm', y='body_mass_g', data=df)
```
